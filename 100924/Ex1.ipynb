```python
import numpy as np
import matplotlib.pyplot as plt
```

Part 1: Without Classes

```python
def plummer_radius(N, a=1.0):
    """
    Generates radii based on the Plummer distribution
    """
    P_r = np.random.random(N)
    r = a * (P_r ** (-2.0/3.0) - 1) ** (-0.5)
    return r

def generate_star_cluster(N, a=1.0):
    """
    Generates star positions in a spherical cluster following Plummer model
    """
    # Generate random r, theta, phi
    r = plummer_radius(N, a)
    theta = np.arccos(2 * np.random.random(N) - 1)  # Uniform in cos(theta)
    phi = 2 * np.pi * np.random.random(N)  # Uniform in phi

    # Convert to Cartesian coordinates
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    return x, y, z

```

```python
# Generate 1000 stars
N = 1000
x, y, z = generate_star_cluster(N)

# Plotting the star cluster
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, s=1)
ax.set_title("Plummer Model Star Cluster")
plt.show()
```

Part 2: With Classes

```python
class StarCluster:
    def __init__(self, N, a=1.0):
        """
        Initializes the star cluster with N stars and scale length a
        """
        self.N = N
        self.a = a
        self.positions = self.generate_star_cluster()

    def plummer_radius(self):
        """
        Generates radii based on the Plummer distribution
        """
        P_r = np.random.random(self.N)
        r = self.a * (P_r ** (-2.0 / 3.0) - 1) ** (-0.5)
        return r

    def generate_star_cluster(self):
        """
        Generates star positions in a spherical cluster following the Plummer model
        """
        r = self.plummer_radius()
        theta = np.arccos(2 * np.random.random(self.N) - 1)  # Uniform in cos(theta)
        phi = 2 * np.pi * np.random.random(self.N)  # Uniform in phi

        x = r * np.sin(theta) * np.cos(phi)
        y = r * np.sin(theta) * np.sin(phi)
        z = r * np.cos(theta)

        return np.vstack((x, y, z)).T
    
    def print_positions(self):
        """
        Prints the positions of the stars in the cluster
        """
        for i, (x, y, z) in enumerate(self.positions):
            print(f"Star {i+1}: x = {x:.3f}, y = {y:.3f}, z = {z:.3f}")
            
    def plot_cluster(self):
        """
        Plots the star cluster in 3D
        """
        fig = plt.figure(figsize=(8, 8))
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter(self.positions[:, 0], self.positions[:, 1], self.positions[:, 2], s=1)
        ax.set_title("Plummer Model Star Cluster")
        plt.show()


```

```python
# Create an instance of the StarCluster class with 1000 stars
cluster = StarCluster(N=1000)

# Print the positions of the stars
cluster.print_positions()

# Optionally, you can still plot the cluster
cluster.plot_cluster()
```
