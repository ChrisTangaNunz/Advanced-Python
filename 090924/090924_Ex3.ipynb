```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
chirp_mass_bins = np.loadtxt('/home/chris/PythonAdvanced/090924/chirpmass_bin.dat')
merger_time_bins = np.loadtxt('/home/chris/PythonAdvanced/090924/tmerg_bin.dat')
chirp_mass_merger_matrix = np.loadtxt('/home/chris/PythonAdvanced/090924/chirpmass_tmerg_tot.dat')

# Transpose the matrix if the shapes are reversed
if chirp_mass_merger_matrix.shape == (201, 1402):
    chirp_mass_merger_matrix = chirp_mass_merger_matrix.T
```

```python
# Create a contour plot
plt.figure(figsize=(8, 6))
cp = plt.contourf(chirp_mass_bins, merger_time_bins, chirp_mass_merger_matrix, levels=20, cmap='viridis')
plt.colorbar(cp)

# Set labels
plt.xlabel('Chirp Mass (M⊙)')
plt.ylabel('Merger Time (Gyr)')
plt.title('Contour Map of Chirp Mass vs Merger Time')

# Show the plot
plt.show()
```
