```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('/home/chris/PythonAdvanced/BBH_first_gen.txt', usecols=(1, 2))  # Read columns 1 and 2 (m1 and m2)
m1 = data[:, 0]  # Primary black hole mass (m1)
m2 = data[:, 1]  # Secondary black hole mass (m2)
```

```python
#Scatter plot m2 vs m1
plt.figure(figsize=(8, 6))
plt.scatter(m1, m2, alpha=0.5)
plt.xlabel('Primary Black Hole Mass (m1) [$M_{\odot}$]')
plt.ylabel('Secondary Black Hole Mass (m2) [$M_{\odot}$]')
plt.title('Scatter plot: m2 vs m1')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist2d(m1, m2, bins=50)  
plt.colorbar(label='Number of Systems')
plt.xlabel('Primary Black Hole Mass (m1) [$M_{\odot}$]')
plt.ylabel('Secondary Black Hole Mass (m2) [$M_{\odot}$]')
plt.title('2D Histogram: m2 vs m1')
plt.grid(True)
plt.show()
```

```python
#D: Calculate mean and standard deviation of m1 and m2
mean_m1 = np.mean(m1)
std_m1 = np.std(m1)
mean_m2 = np.mean(m2)
std_m2 = np.std(m2)

print(f"Mean of m1: {mean_m1:.2f}, Standard Deviation of m1: {std_m1:.2f}")
print(f"Mean of m2: {mean_m2:.2f}, Standard Deviation of m2: {std_m2:.2f}")
```
